Calculate the correlation between vaccination rates and income groups:

```python
seaborn_df['IncomeGroupValue'] = seaborn_df['IncomeGroup'].replace({"High income": 4, "Upper middle income": 3, "Lower middle income": 2, "Low income": 1})
```

```python
from scipy.stats import spearmanr

def correlate_fun(x):
    a = x['IncomeGroupValue'].to_list()
    b = x['value_'].to_list()
    return(spearmanr(a, b)[0])
```

```python
correlation_df = seaborn_df.groupby(['year']).apply(correlate_fun)
corr_df = correlation_df.reset_index()
corr_df.columns = ['year', 'correlation']
corr_df
```
